```python
import pandas as pd

def process_data_with_mapping(mapping_df, data):
    # 创建 DataFrame 格式的数据 df
    df = pd.DataFrame(columns=mapping_df.iloc[:,0].values)
    
    # 填充 df 数据
    for column in df.columns:
        if column in mapping_df.iloc[:,0].values:
            mapping_row = mapping_df[mapping_df.iloc[:,0] == column].iloc[0]
            mapping_name = mapping_row.iloc[1]
            
            if mapping_name in data.columns:
                df[column] = data[mapping_name]
        else:
            df[column] = None
    
    # 找出没有对应关系的列
    dropped_name = [column for column in data.columns if column not in mapping_df.iloc[:,1].values]
    dropped = data[dropped_name]
    
    return df, dropped

def create_mapping_from_excel(filename, sheet_name, Source_ID):
    # 读取 Excel 文件
    df = pd.read_excel(filename, sheet_name = sheet_name)
    mapping = df.loc[df['Source ID'].astype(str).str.contains(str(Source_ID))].iloc[:,[1,4]]
    return mapping
```

```python
# 获取所有sheet的名称
filename = 'Combined Fields V2.0.xlsx'
sheet_names = pd.ExcelFile(filename).sheet_names

# 打印sheet名称
for sheet_name in sheet_names:
    print(sheet_name)
```

```python
#修改sheet_name和Source_ID,可生成对应的关系表
filename = 'Combined Fields V2.0.xlsx'
sheet_name = 'NYC V1'
Source_ID = 1

#调用函数create_mapping_from_excel（）生成关系表
mapping_df = create_mapping_from_excel(filename, sheet_name, Source_ID)
mapping_df
```

```python
#修改文件路径，导入需要修改列名文件的路径
df = pd.read_csv('DOB_Permit_Issuance.csv')
```

```python
#调用函数process_data_with_mapping（），修改列名
data, dropped = process_data_with_mapping(mapping_df, df)
```
